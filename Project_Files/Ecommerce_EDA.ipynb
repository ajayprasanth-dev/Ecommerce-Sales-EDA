```python
df["Order Date"] =pd.to_datetime(df["Order Date"])
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x="Category", data=df)
plt.title("Orders by Category")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x="Region", data=df)
plt.title("Orders by Region")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x="Payment Mode", data=df)
plt.title("Orders by Payment Mode")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(x="Sales", data=df)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(x="Quantity", data=df)
plt.title("Quantity Distribution")
plt.xlabel("Quantity")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="Category", y="Sales",data=df)
plt.title("Category wise Sales Performance")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="Region", y="Profit",data=df)
plt.title("Region wise Profit Performance")
plt.xlabel("Region")
plt.ylabel("Profit")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x="Discount", y="Sales",data=df)
plt.title("Impact of discounts on sale")
plt.xlabel("Discount")
plt.ylabel("Sale")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="Category", y="Profit",data=df)
plt.title("Category wise Profit Performance")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
